```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

#### Deforestation

```python
deforestation_file_path = 'Mapbiomas_Manaus_Deforestation_1986-2023.csv'

deforestation_df = pd.read_csv(deforestation_file_path, sep=';')
deforestation_df
```

```python
deforestation_df = deforestation_df.replace({'-': ''}, regex=True)
deforestation_df
```

```python
deforestation_df.columns
```

```python
selected_columns_2017_2021 = ['Classes', '2017', '2018', '2019', '2020', '2021']

# Create a new DataFrame from the selected columns
deforestation_df_2017_2021 = deforestation_df[selected_columns_2017_2021]

# Display the new DataFrame
deforestation_df_2017_2021
```

```python
deforestation_df_2017_2021.replace('', np.nan, inplace=True)

# Drop rows with NaN values
deforestation_df_2017_2021_dropped = deforestation_df_2017_2021.dropna()
deforestation_df_2017_2021_dropped
```

```python
#### Deforested areas in hectares
```

```python
#### Annual forest suppression and non-forest natural vegetation data generated in Brazil 
#### from the analysis of the history of pixels in the MapBiomas collection of annual land cover 
#### and use maps, with a spatial resolution of 30 m, minimum area of 1 ha and filter time to remove noise.
```

```python
deforestation_df_2017_2021_dropped.set_index('Classes', inplace=True)
```

```python
transposed_deforestation_2017_2021_df = deforestation_df_2017_2021_dropped.T
transposed_deforestation_2017_2021_df
```

```python
numeric_deforestation_2017_2021_df = transposed_deforestation_2017_2021_df.astype(int)
numeric_deforestation_2017_2021_df = numeric_deforestation_2017_2021_df.reset_index()
numeric_deforestation_2017_2021_df.rename(columns={'index': 'YEAR_NOTIF'}, inplace=True)
numeric_deforestation_2017_2021_df
```

```python
numeric_deforestation_2017_2021_df.plot(kind='line', x='YEAR_NOTIF', y='2. Supressão de Vegetação Secundária');
plt.show()
```

```python
cols_to_convert = numeric_deforestation_2017_2021_df.columns.difference(['Classes', 'YEAR_NOTIF'])
numeric_deforestation_2017_2021_df[cols_to_convert] = numeric_deforestation_2017_2021_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
numeric_deforestation_2017_2021_df
```

```python
# Convert the index to datetime format for yearly data
monthly_data = numeric_deforestation_2017_2021_df[cols_to_convert].values.repeat(12, axis=0) / 12

# Create a new DataFrame with monthly data, assigning a monthly date range as the index
monthly_index = pd.date_range(start='2017-01-01', end='2021-12-31', freq='M')
monthly_deforestation_2017_2021_df = pd.DataFrame(monthly_data, index=monthly_index, columns=cols_to_convert)

# Plot the '2. Supressão de Vegetação Secundária' column to confirm it has no NaNs and data looks correct
monthly_deforestation_2017_2021_df['2. Supressão de Vegetação Secundária'].plot(kind='line', figsize=(10, 6))
plt.title('Monthly Deforestation - 2. Supressão de Vegetação Secundária')
plt.xlabel('Date')
plt.ylabel('Deforestation (hectares)')
plt.grid(True)
plt.show()

# Display the final DataFrame to confirm it covers the full date range
monthly_deforestation_2017_2021_df = monthly_deforestation_2017_2021_df.reset_index()
monthly_deforestation_2017_2021_df.rename(columns={'index': 'MONTH_NOTIF'}, inplace=True)
monthly_deforestation_2017_2021_df
```

```python
monthly_deforestation_2017_2021_df['MONTH_NOTIF'] = pd.to_datetime(monthly_deforestation_2017_2021_df['MONTH_NOTIF']).dt.to_period('M').dt.to_timestamp()
monthly_deforestation_2017_2021_df
```

```python
monthly_deforestation_2017_2021_df.to_csv('manaus_monthly_deforestation.csv', sep=',', index=False, encoding='utf-8')
numeric_deforestation_2017_2021_df.to_csv('manaus_yearly_deforestation.csv', sep=',', index=False, encoding='utf-8')
```

#### Forest Fires

```python
forest_fires_file_path = 'Mapbiomas_Manaus_Forest_Fires_1986-2023.csv'

forest_fires_df = pd.read_csv(forest_fires_file_path, sep=';')
forest_fires_df
```

```python
forest_fires_df = forest_fires_df.replace({'-': ''}, regex=True)
forest_fires_df
```

```python
forest_fires_df.columns
```

```python
# Create a new DataFrame from the selected columns
forest_fires_2017_2021_df = forest_fires_df[selected_columns_2017_2021]

# Display the new DataFrame
forest_fires_2017_2021_df
```

```python
forest_fires_2017_2021_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
numeric_columns = ['2017', '2018', '2019', '2020', '2021']

# Drop rows where all values in the numeric columns are NaN
forest_fires_df_2017_2021_dropped = forest_fires_2017_2021_df.dropna(subset=numeric_columns, how='all')
forest_fires_df_2017_2021_dropped = forest_fires_df_2017_2021_dropped.fillna(0)
forest_fires_df_2017_2021_dropped
```

```python
#### Burnt out areas in hectares
```

```python
#### Historical series, mapped since 1985 and updated annually, which presents the 
#### scars of fire over the national territory over 39 years (1985 to 2023).
#### All mapping of fire scars in Brazil was based on image mosaics from Landsat 
#### satellites with a spatial resolution of 30 meters, covering the entire Brazilian territory.
```

```python
forest_fires_df_2017_2021_dropped.set_index('Classes', inplace=True)
```

```python
transposed_forest_fires_2017_2021_df = forest_fires_df_2017_2021_dropped.T
transposed_forest_fires_2017_2021_df
```

```python
numeric_forest_fires_2017_2021_df = transposed_forest_fires_2017_2021_df.astype(int)
numeric_forest_fires_2017_2021_df
```

```python
numeric_forest_fires_2017_2021_df.plot(kind='line', y='1. Floresta');
plt.show()
```

```python
numeric_forest_fires_2017_2021_df = numeric_forest_fires_2017_2021_df.assign(Total = (numeric_forest_fires_2017_2021_df.get('1. Floresta')) + 
                                                                 (numeric_forest_fires_2017_2021_df.get('2. Formação Natural não Florestal')) +
                                                                 (numeric_forest_fires_2017_2021_df.get('3. Agropecuária')))
numeric_forest_fires_2017_2021_df
```

```python
numeric_forest_fires_2017_2021_df = numeric_forest_fires_2017_2021_df.reset_index()
numeric_forest_fires_2017_2021_df.rename(columns={'index': 'YEAR_NOTIF'}, inplace=True)
numeric_forest_fires_2017_2021_df
```

```python
numeric_forest_fires_2017_2021_df.plot(kind='line', x='YEAR_NOTIF', y='Total');
plt.show()
```

#### Monthly forest fires

```python
monthly_forest_fires_file_path = 'Mapbiomas_Monthly_Manaus_Forest_Fires_1986-2023.csv'

monthly_forest_fires_df = pd.read_csv(monthly_forest_fires_file_path, sep=';')

#monthly_forest_fires_df.set_index('Classes', inplace=True)

months = ['January', 'February', 'March', 'April', 'May', 
               'June', 'July', 'August', 'September', 'October', 'November', 'December']

monthly_forest_fires_df['Classes'] = monthly_forest_fires_df['Classes'].replace({'Janeiro': months[0], 
                                                                                 'Fevereiro': months[1],
                                                                                 'Março': months[2],
                                                                                 'Abril': months[3],
                                                                                 'Maio': months[4],
                                                                                 'Junho': months[5],
                                                                                 'Julho': months[6],
                                                                                 'Agosto': months[7],
                                                                                 'Setembro': months[8],
                                                                                 'Outubro': months[9],
                                                                                 'Novembro': months[10],
                                                                                 'Dezembro': months[11]})

# monthly_forest_fires_df.columns = month_names

monthly_forest_fires_df
```

```python
monthly_forest_fires_df = monthly_forest_fires_df.replace({'-': ''}, regex=True)
monthly_forest_fires_df
```

```python
# Create a new DataFrame from the selected columns
monthly_forest_fires_2017_2021_df = monthly_forest_fires_df[selected_columns_2017_2021]

# Display the new DataFrame
monthly_forest_fires_2017_2021_df
```

```python
monthly_forest_fires_2017_2021_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
numeric_columns = ['2017', '2018', '2019', '2020', '2021']

# Drop rows where all values in the numeric columns are NaN
monthly_forest_fires_df_2017_2021_dropped = monthly_forest_fires_2017_2021_df.dropna(subset=numeric_columns, how='all')
monthly_forest_fires_df_2017_2021_dropped = monthly_forest_fires_df_2017_2021_dropped.fillna(0)
monthly_forest_fires_df_2017_2021_dropped
```

```python
monthly_forest_fires_df_2017_2021_dropped.set_index('Classes', inplace=True)
transposed_monthly_forest_fires_2017_2021_df = monthly_forest_fires_df_2017_2021_dropped.T
transposed_monthly_forest_fires_2017_2021_df
```

```python
numeric_monthly_forest_fires_2017_2021_df = transposed_monthly_forest_fires_2017_2021_df.astype(int)
numeric_monthly_forest_fires_2017_2021_df
```

```python
plt.figure(figsize=(12, 6))
numeric_monthly_forest_fires_2017_2021_df.plot(kind='line')

# Title and axis labels
plt.title('Monthly Forest Fires (2017-2021)')
plt.xlabel('Months')
plt.ylabel('Burnt Out Area (Ha)')

# Adjusting the x-ticks rotation
plt.xticks(rotation=45)

# Moving the legend outside the plot
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding grid
plt.grid(True)

# Display the plot
plt.tight_layout()  # Adjust the layout so everything fits
plt.show()
```

```python
numeric_monthly_forest_fires_2017_2021_df
```

```python
monthly_forest_fires_df_2017_2021_dropped['Month'] = months  # This adds months as index
long_forest_fires_2017_2021_df = monthly_forest_fires_df_2017_2021_dropped.melt(id_vars='Month', var_name='Year', value_name='Fires')

long_forest_fires_2017_2021_df['Year'] = long_forest_fires_2017_2021_df['Year'].astype(str)
long_forest_fires_2017_2021_df['Fires'] = long_forest_fires_2017_2021_df['Fires'].astype(int)

long_forest_fires_2017_2021_df['Date'] = pd.to_datetime(long_forest_fires_2017_2021_df['Year'] + '-' + long_forest_fires_2017_2021_df['Month'], format='%Y-%B')

plt.figure(figsize=(12, 6))
plt.plot(long_forest_fires_2017_2021_df['Date'], long_forest_fires_2017_2021_df['Fires'], marker='o')

# Adding labels and title
plt.title('Monthly Forest Fires')
plt.xlabel('Date')
plt.ylabel('Burnt Out Area (Ha)')

# Format the x-axis to show the dates clearly
plt.xticks(rotation=45)
plt.grid(True)

# Adjust layout to fit everything
plt.tight_layout()

# Display the plot
plt.show()
```

```python
numeric_monthly_forest_fires_2017_2021_df = numeric_monthly_forest_fires_2017_2021_df.reset_index()
numeric_monthly_forest_fires_2017_2021_df.rename(columns={'index': 'YEAR_NOTIF'}, inplace=True)
numeric_monthly_forest_fires_2017_2021_df
```

```python
numeric_monthly_forest_fires_2017_2021_long_df = numeric_monthly_forest_fires_2017_2021_df.melt(id_vars='YEAR_NOTIF', var_name='Month', value_name='Total')

# Map month names to numbers for sorting and formatting
month_map = {
    'January': '01', 'February': '02', 'March': '03', 'April': '04',
    'May': '05', 'June': '06', 'July': '07', 'August': '08', 
    'September': '09', 'October': '10', 'November': '11', 'December': '12'
}

numeric_monthly_forest_fires_2017_2021_long_df['Month'] = numeric_monthly_forest_fires_2017_2021_long_df['Month'].map(month_map)

# Create the MONTH_NOTIF column in the desired format
numeric_monthly_forest_fires_2017_2021_long_df['MONTH_NOTIF'] = pd.to_datetime(numeric_monthly_forest_fires_2017_2021_long_df['YEAR_NOTIF'].astype(str) + '-' + numeric_monthly_forest_fires_2017_2021_long_df['Month'] + '-01')

# Drop the Year and Month columns and sort the DataFrame
numeric_monthly_forest_fires_2017_2021_long_df = numeric_monthly_forest_fires_2017_2021_long_df.drop(columns=['YEAR_NOTIF', 'Month']).sort_values('MONTH_NOTIF').reset_index(drop=True)
numeric_monthly_forest_fires_2017_2021_long_df
```

```python
cols = numeric_monthly_forest_fires_2017_2021_long_df.columns.tolist()
cols
```

```python
cols = cols[-1:] + cols[:-1]
numeric_monthly_forest_fires_2017_2021_long_df = numeric_monthly_forest_fires_2017_2021_long_df[cols]
numeric_monthly_forest_fires_2017_2021_long_df
```

```python
numeric_monthly_forest_fires_2017_2021_long_df['MONTH_NOTIF'] = pd.to_datetime(numeric_monthly_forest_fires_2017_2021_long_df['MONTH_NOTIF'])

plt.figure(figsize=(12, 6))
plt.plot(numeric_monthly_forest_fires_2017_2021_long_df['MONTH_NOTIF'], numeric_monthly_forest_fires_2017_2021_long_df['Total'], marker='o')

plt.title('Total Burnt Area (ha)')
plt.xlabel('Month')
plt.ylabel('Total')
plt.xticks(rotation=45)  
plt.grid(True)  

plt.tight_layout()
plt.show()
```

```python
numeric_monthly_forest_fires_2017_2021_long_df.to_csv('manaus_monthly_forest_fires.csv', sep=',', index=False, encoding='utf-8')
numeric_forest_fires_2017_2021_df.to_csv('manaus_yearly_forest_fires.csv', sep=',', index=False, encoding='utf-8')
```
