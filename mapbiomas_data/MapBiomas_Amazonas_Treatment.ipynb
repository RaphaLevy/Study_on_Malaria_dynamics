```python
with open('Amazonas-MapBiomas - Tabela de Dados.csv', 'r', encoding='utf-8') as f:
    content = f.read()

# Step 2: Replace any extra quotes if needed and split the text by commas
content = content.replace('"', '')

# Step 3: Use StringIO to convert the string into a file-like object
from io import StringIO
data = StringIO(content)

# Step 4: Read the data using pandas, treating commas as the delimiter
df = pd.read_csv(data, delimiter=',')

# Step 5: Display the DataFrame to confirm it has been correctly split into columns
df
```
