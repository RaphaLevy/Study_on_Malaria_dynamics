```python
amazonas_deforestation_index_reset_df_safe = amazonas_deforestation_index_reset_df.replace(0, np.nan)
amazonas_deforestation_index_reset_df_safe
```

```python
ratio = manaus_deforestation_index_reset_df / amazonas_deforestation_index_reset_df_safe * 100
ratio
```

```python
ratio = ratio.reset_index()
ratio
```

```python
selected_columns_2017_2021 = ['Classes', '2017', '2018', '2019', '2020', '2021']

# Create a new DataFrame from the selected columns
ratio_df_2017_2021 = ratio[selected_columns_2017_2021]

# Display the new DataFrame
ratio_df_2017_2021
```

```python
# Drop rows with NaN values
ratio_df_2017_2021_dropped = ratio_df_2017_2021.dropna()
ratio_df_2017_2021_dropped
```

```python
ratio_df_2017_2021_dropped.set_index('Classes', inplace=True)
```

```python
ratio_df_2017_2021_dropped
```

```python
non_zero_ratio = ratio_df_2017_2021_dropped[ratio_df_2017_2021_dropped > 0]
non_zero_ratio
```

```python
# Drop rows with NaN values
non_zero_ratio = non_zero_ratio.dropna()
non_zero_ratio
```

```python
transposed_ratio_2017_2021_df = non_zero_ratio.T
transposed_ratio_2017_2021_df
```
