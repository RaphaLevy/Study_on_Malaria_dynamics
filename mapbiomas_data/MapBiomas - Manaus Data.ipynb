```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

#### Deforestation

```python
deforestation_file_path = 'Mapbiomas_Manaus_Deforestation_1986-2023.csv'

deforestation_df = pd.read_csv(deforestation_file_path, sep=';')
deforestation_df
```

```python
deforestation_df = deforestation_df.replace({'-': ''}, regex=True)
deforestation_df
```

```python
deforestation_df.columns
```

```python
selected_columns = ['Classes', '2004', '2005', '2006', '2007', '2008']

# Create a new DataFrame from the selected columns
new_deforestation_df = deforestation_df[selected_columns]

# Display the new DataFrame
new_deforestation_df
```

```python
new_deforestation_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
deforestation_df_dropped = new_deforestation_df.dropna()
deforestation_df_dropped
```

```python
#### Deforested areas in hectares
```

```python
#### Annual forest suppression and non-forest natural vegetation data generated in Brazil 
#### from the analysis of the history of pixels in the MapBiomas collection of annual land cover 
#### and use maps, with a spatial resolution of 30 m, minimum area of 1 ha and filter time to remove noise.
```

```python
deforestation_df_dropped.set_index('Classes', inplace=True)
```

```python
transposed_deforestation_df = deforestation_df_dropped.T
transposed_deforestation_df
```

```python
numeric_deforestation_df = transposed_deforestation_df.astype(int)
numeric_deforestation_df
```

```python
numeric_deforestation_df.plot(kind='line', y='2. Supressão de Vegetação Secundária');
plt.show()
```

#### Forest Fires

```python
forest_fires_file_path = 'Mapbiomas_Manaus_Forest_Fires_1986-2023.csv'

forest_fires_df = pd.read_csv(forest_fires_file_path, sep=';')
forest_fires_df
```

```python
forest_fires_df = forest_fires_df.replace({'-': ''}, regex=True)
forest_fires_df
```

```python
forest_fires_df.columns
```

```python
# Create a new DataFrame from the selected columns
new_forest_fires_df = forest_fires_df[selected_columns]

# Display the new DataFrame
new_forest_fires_df
```

```python
new_forest_fires_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
numeric_columns = ['2004', '2005', '2006', '2007', '2008']

# Drop rows where all values in the numeric columns are NaN
forest_fires_df_dropped = new_forest_fires_df.dropna(subset=numeric_columns, how='all')
forest_fires_df_dropped = forest_fires_df_dropped.fillna(0)
forest_fires_df_dropped
```

```python
#### Burnt out areas in hectares
```

```python
#### Historical series, mapped since 1985 and updated annually, which presents the 
#### scars of fire over the national territory over 39 years (1985 to 2023).
#### All mapping of fire scars in Brazil was based on image mosaics from Landsat 
#### satellites with a spatial resolution of 30 meters, covering the entire Brazilian territory.
```

```python
forest_fires_df_dropped.set_index('Classes', inplace=True)
```

```python
transposed_forest_fires_df = forest_fires_df_dropped.T
transposed_forest_fires_df
```

```python
numeric_forest_fires_df = transposed_forest_fires_df.astype(int)
numeric_forest_fires_df
```

```python
numeric_forest_fires_df.plot(kind='line', y='1. Floresta');
plt.show()
```

```python
numeric_forest_fires_df = numeric_forest_fires_df.assign(Total = (numeric_forest_fires_df.get('1. Floresta')) + 
                                                                 (numeric_forest_fires_df.get('2. Formação Natural não Florestal')) +
                                                                 (numeric_forest_fires_df.get('3. Agropecuária')))
numeric_forest_fires_df
```

```python
numeric_forest_fires_df.plot(kind='line', y='Total');
plt.show()
```

#### Monthly forest fires

```python
monthly_forest_fires_file_path = 'Mapbiomas_Monthly_Manaus_Forest_Fires_1986-2023.csv'

monthly_forest_fires_df = pd.read_csv(monthly_forest_fires_file_path, sep=';')

#monthly_forest_fires_df.set_index('Classes', inplace=True)

month_names = ['January', 'February', 'March', 'April', 'May', 
               'June', 'July', 'August', 'September', 'October', 'November', 'December']

monthly_forest_fires_df['Classes'] = monthly_forest_fires_df['Classes'].replace({'Janeiro': month_names[0], 
                                                                                 'Fevereiro': month_names[1],
                                                                                 'Março': month_names[2],
                                                                                 'Abril': month_names[3],
                                                                                 'Maio': month_names[4],
                                                                                 'Junho': month_names[5],
                                                                                 'Julho': month_names[6],
                                                                                 'Agosto': month_names[7],
                                                                                 'Setembro': month_names[8],
                                                                                 'Outubro': month_names[9],
                                                                                 'Novembro': month_names[10],
                                                                                 'Dezembro': month_names[11]})

# monthly_forest_fires_df.columns = month_names

monthly_forest_fires_df
```

```python
monthly_forest_fires_df = monthly_forest_fires_df.replace({'-': ''}, regex=True)
monthly_forest_fires_df
```

```python
# Create a new DataFrame from the selected columns
new_monthly_forest_fires_df = monthly_forest_fires_df[selected_columns]

# Display the new DataFrame
new_monthly_forest_fires_df
```

```python
new_monthly_forest_fires_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
numeric_columns = ['2004', '2005', '2006', '2007', '2008']

# Drop rows where all values in the numeric columns are NaN
monthly_forest_fires_df_dropped = new_monthly_forest_fires_df.dropna(subset=numeric_columns, how='all')
monthly_forest_fires_df_dropped = monthly_forest_fires_df_dropped.fillna(0)
monthly_forest_fires_df_dropped
```

```python
monthly_forest_fires_df_dropped.set_index('Classes', inplace=True)
transposed_monthly_forest_fires_df = monthly_forest_fires_df_dropped.T
transposed_monthly_forest_fires_df
```

```python
numeric_monthly_forest_fires_df = transposed_monthly_forest_fires_df.astype(int)
numeric_monthly_forest_fires_df
```

```python
plt.figure(figsize=(12, 6))
numeric_monthly_forest_fires_df.plot(kind='line')
plt.title('Monthly Forest Fires (2004-2008)')
plt.xlabel('Months')
plt.ylabel('Number of Forest Fires')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.grid(True)
plt.show()
```
