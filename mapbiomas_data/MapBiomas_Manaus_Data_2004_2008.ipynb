```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

#### Deforestation

```python
deforestation_file_path = 'Mapbiomas_Manaus_Deforestation_1986-2023.csv'

deforestation_df = pd.read_csv(deforestation_file_path, sep=';')
deforestation_df
```

```python
deforestation_df = deforestation_df.replace({'-': ''}, regex=True)
deforestation_df
```

```python
deforestation_df.columns
```

```python
selected_columns_2004_2008 = ['Classes', '2004', '2005', '2006', '2007', '2008']

# Create a new DataFrame from the selected columns
deforestation_df_2004_2008 = deforestation_df[selected_columns_2004_2008]

# Display the new DataFrame
deforestation_df_2004_2008
```

```python
deforestation_df_2004_2008.replace('', np.nan, inplace=True)

# Drop rows with NaN values
deforestation_df_2004_2008_dropped = deforestation_df_2004_2008.dropna()
deforestation_df_2004_2008_dropped
```

```python
#### Deforested areas in hectares
```

```python
#### Annual forest suppression and non-forest natural vegetation data generated in Brazil 
#### from the analysis of the history of pixels in the MapBiomas collection of annual land cover 
#### and use maps, with a spatial resolution of 30 m, minimum area of 1 ha and filter time to remove noise.
```

```python
deforestation_df_2004_2008_dropped.set_index('Classes', inplace=True)
```

```python
transposed_deforestation_2004_2008_df = deforestation_df_2004_2008_dropped.T
transposed_deforestation_2004_2008_df
```

```python
numeric_deforestation_2004_2008_df = transposed_deforestation_2004_2008_df.astype(int)
numeric_deforestation_2004_2008_df
```

```python
numeric_deforestation_2004_2008_df.plot(kind='line', y='2. Supressão de Vegetação Secundária');
plt.show()
```

```python
# Convert the index to datetime format for yearly data
monthly_data = numeric_deforestation_2004_2008_df.values.repeat(12, axis=0) / 12

# Create a new DataFrame with monthly data, assigning a monthly date range as the index
monthly_index = pd.date_range(start='2004-01-01', end='2008-12-31', freq='M')
monthly_deforestation_2004_2008_df = pd.DataFrame(monthly_data, index=monthly_index, columns=numeric_deforestation_2004_2008_df.columns)

# Plot the '2. Supressão de Vegetação Secundária' column to confirm it has no NaNs and data looks correct
monthly_deforestation_2004_2008_df['2. Supressão de Vegetação Secundária'].plot(kind='line', figsize=(10, 6))
plt.title('Monthly Deforestation - 2. Supressão de Vegetação Secundária')
plt.xlabel('Date')
plt.ylabel('Deforestation (hectares)')
plt.grid(True)
plt.show()

# Display the final DataFrame to confirm it covers the full date range
monthly_deforestation_2004_2008_df
```

#### Forest Fires

```python
forest_fires_file_path = 'Mapbiomas_Manaus_Forest_Fires_1986-2023.csv'

forest_fires_df = pd.read_csv(forest_fires_file_path, sep=';')
forest_fires_df
```

```python
forest_fires_df = forest_fires_df.replace({'-': ''}, regex=True)
forest_fires_df
```

```python
forest_fires_df.columns
```

```python
# Create a new DataFrame from the selected columns
forest_fires_2004_2008_df = forest_fires_df[selected_columns_2004_2008]

# Display the new DataFrame
forest_fires_2004_2008_df
```

```python
forest_fires_2004_2008_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
numeric_columns = ['2004', '2005', '2006', '2007', '2008']

# Drop rows where all values in the numeric columns are NaN
forest_fires_df_2004_2008_dropped = forest_fires_2004_2008_df.dropna(subset=numeric_columns, how='all')
forest_fires_df_2004_2008_dropped = forest_fires_df_2004_2008_dropped.fillna(0)
forest_fires_df_2004_2008_dropped
```

```python
#### Burnt out areas in hectares
```

```python
#### Historical series, mapped since 1985 and updated annually, which presents the 
#### scars of fire over the national territory over 39 years (1985 to 2023).
#### All mapping of fire scars in Brazil was based on image mosaics from Landsat 
#### satellites with a spatial resolution of 30 meters, covering the entire Brazilian territory.
```

```python
forest_fires_df_2004_2008_dropped.set_index('Classes', inplace=True)
```

```python
transposed_forest_fires_2004_2008_df = forest_fires_df_2004_2008_dropped.T
transposed_forest_fires_2004_2008_df
```

```python
numeric_forest_fires_2004_2008_df = transposed_forest_fires_2004_2008_df.astype(int)
numeric_forest_fires_2004_2008_df
```

```python
numeric_forest_fires_2004_2008_df.plot(kind='line', y='1. Floresta');
plt.show()
```

```python
numeric_forest_fires_2004_2008_df = numeric_forest_fires_2004_2008_df.assign(Total = (numeric_forest_fires_2004_2008_df.get('1. Floresta')) + 
                                                                 (numeric_forest_fires_2004_2008_df.get('2. Formação Natural não Florestal')) +
                                                                 (numeric_forest_fires_2004_2008_df.get('3. Agropecuária')))
numeric_forest_fires_2004_2008_df
```

```python
numeric_forest_fires_2004_2008_df.plot(kind='line', y='Total');
plt.show()
```

#### Monthly forest fires

```python
monthly_forest_fires_file_path = 'Mapbiomas_Monthly_Manaus_Forest_Fires_1986-2023.csv'

monthly_forest_fires_df = pd.read_csv(monthly_forest_fires_file_path, sep=';')

#monthly_forest_fires_df.set_index('Classes', inplace=True)

months = ['January', 'February', 'March', 'April', 'May', 
               'June', 'July', 'August', 'September', 'October', 'November', 'December']

monthly_forest_fires_df['Classes'] = monthly_forest_fires_df['Classes'].replace({'Janeiro': months[0], 
                                                                                 'Fevereiro': months[1],
                                                                                 'Março': months[2],
                                                                                 'Abril': months[3],
                                                                                 'Maio': months[4],
                                                                                 'Junho': months[5],
                                                                                 'Julho': months[6],
                                                                                 'Agosto': months[7],
                                                                                 'Setembro': months[8],
                                                                                 'Outubro': months[9],
                                                                                 'Novembro': months[10],
                                                                                 'Dezembro': months[11]})

# monthly_forest_fires_df.columns = month_names

monthly_forest_fires_df
```

```python
monthly_forest_fires_df = monthly_forest_fires_df.replace({'-': ''}, regex=True)
monthly_forest_fires_df
```

```python
# Create a new DataFrame from the selected columns
monthly_forest_fires_2004_2008_df = monthly_forest_fires_df[selected_columns_2004_2008]

# Display the new DataFrame
monthly_forest_fires_2004_2008_df
```

```python
monthly_forest_fires_2004_2008_df.replace('', np.nan, inplace=True)

# Drop rows with NaN values
numeric_columns = ['2004', '2005', '2006', '2007', '2008']

# Drop rows where all values in the numeric columns are NaN
monthly_forest_fires_df_2004_2008_dropped = monthly_forest_fires_2004_2008_df.dropna(subset=numeric_columns, how='all')
monthly_forest_fires_df_2004_2008_dropped = monthly_forest_fires_df_2004_2008_dropped.fillna(0)
monthly_forest_fires_df_2004_2008_dropped
```

```python
monthly_forest_fires_df_2004_2008_dropped.set_index('Classes', inplace=True)
transposed_monthly_forest_fires_2004_2008_df = monthly_forest_fires_df_2004_2008_dropped.T
transposed_monthly_forest_fires_2004_2008_df
```

```python
numeric_monthly_forest_fires_2004_2008_df = transposed_monthly_forest_fires_2004_2008_df.astype(int)
numeric_monthly_forest_fires_2004_2008_df
```

```python
plt.figure(figsize=(12, 6))
numeric_monthly_forest_fires_2004_2008_df.plot(kind='line')

# Title and axis labels
plt.title('Monthly Forest Fires (2004-2008)')
plt.xlabel('Months')
plt.ylabel('Number of Forest Fires')

# Adjusting the x-ticks rotation
plt.xticks(rotation=45)

# Moving the legend outside the plot
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding grid
plt.grid(True)

# Display the plot
plt.tight_layout()  # Adjust the layout so everything fits
plt.show()
```

```python
numeric_monthly_forest_fires_2004_2008_df
```

```python
monthly_forest_fires_df_2004_2008_dropped['Month'] = months  # This adds months as index
long_forest_fires_2004_2008_df = monthly_forest_fires_df_2004_2008_dropped.melt(id_vars='Month', var_name='Year', value_name='Fires')

long_forest_fires_2004_2008_df['Year'] = long_forest_fires_2004_2008_df['Year'].astype(str)
long_forest_fires_2004_2008_df['Fires'] = long_forest_fires_2004_2008_df['Fires'].astype(int)

long_forest_fires_2004_2008_df['Date'] = pd.to_datetime(long_forest_fires_2004_2008_df['Year'] + '-' + long_forest_fires_2004_2008_df['Month'], format='%Y-%B')

plt.figure(figsize=(12, 6))
plt.plot(long_forest_fires_2004_2008_df['Date'], long_forest_fires_2004_2008_df['Fires'], marker='o')

# Adding labels and title
plt.title('Monthly Forest Fires')
plt.xlabel('Date')
plt.ylabel('Number of Forest Fires')

# Format the x-axis to show the dates clearly
plt.xticks(rotation=45)
plt.grid(True)

# Adjust layout to fit everything
plt.tight_layout()

# Display the plot
plt.show()
```
