```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from numpy import cos
from math import log
from scipy.interpolate import interp1d
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize_scalar
from scipy.optimize import minimize
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from sage.plot.plot3d.parametric_surface import ParametricSurface
import plotly.graph_objects as go
import plotly.express as px
from scipy.optimize import fsolve
from sympy import symbols, Eq, solve
```

```python
T_prime=19.9

T1=23.2
T2=0.07
omega1=0.67
phi1=1.53

R1=85.9
R2=0.98
omega2=0.65
phi2=1.99

B_E=200
p_ME=0.9
p_ML=0.25
p_MP=0.75
tau_E=1
tau_P=1
c1=0.00554
c2=-0.06737
D1=36.5
b1=0.04
A=-0.03
B=1.31
C=-4.4
b2=0.09
tau_H=10
DD=105
Tmin=14.5
gamma= 1/120
R_L = 50
N=8558
M= 10000  #0
E_M0=1 #0 # None
I_H0=1 
I_M0=0
R_H0=0

if E_M0 is None:
        E_M0 = M / 100

S_H0 = N - I_H0 - R_H0
S_M0 = M - E_M0 - I_M0
```

```python
def original_model_with_derivatives(t, z, T1, T2, omega1, phi1, R1, R2, omega2, phi2,
                                    B_E, p_ME, p_ML, p_MP, tau_E, tau_P, c1, c2,
                                    D1, b1, A, B, C, b2, tau_H, DD, Tmin, gamma, R_L,
                                    N, M, I_M0, I_H0):
    
    S_H, I_H, R_H, S_M, E_M, I_M = z
    
    # Temperature and rain functions
    def Temp(t):
        return T1 * (1 + T2 * np.cos(omega1 * t - phi1))

    def Rain(t):
        return R1 * (1 + R2 * np.cos(omega2 * t - phi2))

    R_curr = Rain(t)
    T_curr = Temp(t)
    
    # Duration of larval development phase
    def tau_L(Temp):
        return 1 / (c1 * Temp + c2)
    
    # Daily survival rate of mosquitoes
    def p(Temp):
        return np.exp(-1 / (A * Temp**2 + B * Temp + C))
    
    # Larval survival probability dependent on temperature and rain
    def p_LT(Temp):
        return np.exp(-(c1 * Temp + c2))
    
    def p_LR(Rain):
        return (4 * p_ML / R_L**2) * Rain * (R_L - Rain)
    
    def p_LRT(Rain, Temp):
        return p_LR(Rain) * p_LT(Temp)
    
    # Egg and pupa survival probability dependent on rain
    def p_ER(Rain):
        return (4 * p_ME / R_L**2) * Rain * (R_L - Rain)
    
    def p_PR(Rain):
        return (4 * p_MP / R_L**2) * Rain * (R_L - Rain)

    # Duration of the sporogonic cycle
    def tau_M(Temp):
        return DD / (Temp - Tmin)
    
    tau_M_curr = tau_M(T_curr)

    # Defining the mosquito survival function
    def l_tau_M(tau_M, Temp):  
        p_T = p(Temp)
        return p_T**tau_M
    
    l_curr = l_tau_M(tau_M_curr, T_curr)

    # Defining the mosquito mortality rate
    def mu(Temp):
        return -np.log(p(Temp))
    
    mu_curr = mu(T_curr)
    
    # Calculate the biting rate and transmission probability
    def a(Temp):
        return (T_curr - T_prime) / D1
    
    a_curr = a(T_curr)
    
    def b(Rain, Temp):
        return B_E * p_ER(R_curr) * p_LRT(R_curr, T_curr) * p_PR(R_curr) / (tau_E + tau_L(T_curr) + tau_P)
    
    b_curr = b(R_curr, T_curr)
    
    # Calculate derivatives
    dShdt = -a_curr * b2 * (I_M / N) * S_H
    dIhdt = a_curr * b2 * (I_M / N) * S_H - gamma * I_H
    dRhdt = gamma * I_H
    
    dSmdt = b_curr - a_curr * b1 * (I_H0 / N) * S_M - mu_curr * S_M
    dEmdt = a_curr * b1 * (I_H0 / N) * S_M - mu_curr * E_M - a_curr * b1 * (I_H0 / N) * S_M * l_curr
    dImdt = a_curr * b1 * (I_H0 / N) * S_M * l_curr - mu_curr * I_M
    
    return [dShdt, dIhdt, dRhdt, dSmdt, dEmdt, dImdt]
```

```python
args = (T1, T2, omega1, phi1, R1, R2, omega2, phi2,
                   B_E, p_ME, p_ML, p_MP, tau_E, tau_P, c1, c2,
                   D1, b1, A, B, C, b2, tau_H, DD, Tmin, gamma, R_L,
                   N, M, I_M0, I_H0)
```

```python
t_eval = np.linspace(0, 365, 1000)  # Evaluate the solution at 1000 points between 0 and 365
sol = solve_ivp(original_model_with_derivatives, [0, 365], [S_H0, I_H0, R_H0, S_M0, E_M0, I_M0], args=args, dense_output=True, t_eval=t_eval, method="DOP853")

# Extracting the solution
times = sol.t
S_H, I_H, R_H, S_M, E_M, I_M = sol.y

# Calculate derivatives at each time step
dS_M_dt, dE_M_dt, dI_M_dt = [], [], []
for i in range(len(times)):
    dSh, dIh, dRh, dSmdt, dEmdt, dImdt = original_model_with_derivatives(times[i], [S_H[i], I_H[i], R_H[i], S_M[i], E_M[i], I_M[i]], *args)
    dS_M_dt.append(dSmdt)
    dE_M_dt.append(dEmdt)
    dI_M_dt.append(dImdt)
```

```python
# Plot the population functions
plt.figure(figsize=(10, 6))
plt.plot(times, S_M, label='Susceptible Mosquitoes')
plt.plot(times, E_M, label='Exposed Mosquitoes')
plt.plot(times, I_M, label='Infected Mosquitoes')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.title('Mosquito Populations Over Time')
plt.legend()
plt.show()
```

```python
# Plot the derivatives
plt.figure(figsize=(10, 6))
plt.plot(times, dS_M_dt, label='dS_M/dt (Susceptible)')
plt.plot(times, dE_M_dt, label='dE_M/dt (Exposed)')
plt.plot(times, dI_M_dt, label='dI_M/dt (Infected)')
plt.xlabel('Time (days)')
plt.ylabel('Rate of Change')
plt.title('Derivatives of Mosquito Populations Over Time')
plt.legend()
plt.show()
```
