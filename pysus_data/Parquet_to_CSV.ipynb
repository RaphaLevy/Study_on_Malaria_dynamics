```python
pop_2004_df = pd.read_csv('pop_data/POPTBR04.csv')
pop_2004_df
```

```python
pop_2005_df = pd.read_csv('pop_data/POPTBR05.csv')
pop_2006_df = pd.read_csv('pop_data/POPTBR06.csv')
pop_2007_df = pd.read_csv('pop_data/POPTBR07.csv')
pop_2008_df = pd.read_csv('pop_data/POPTBR08.csv')
pop_2009_df = pd.read_csv('pop_data/POPTBR09.csv')
pop_2010_df = pd.read_csv('pop_data/POPTBR10.csv')
pop_2011_df = pd.read_csv('pop_data/POPTBR11.csv')
pop_2012_df = pd.read_csv('pop_data/POPTBR12.csv')
pop_2013_df = pd.read_csv('pop_data/POPTBR13.csv')
```

```python
full_pop_df = pd.concat([pop_2004_df, pop_2005_df, pop_2006_df,
                     pop_2007_df, pop_2008_df, pop_2009_df,
                     pop_2010_df, pop_2011_df, pop_2012_df, pop_2013_df])
full_pop_df = full_pop_df.rename(columns={"MUNIC_RES": 'ID_MUNICIP'})
full_pop_df = full_pop_df.reset_index()
full_pop_df = full_pop_df.drop(columns=['index'])
full_pop_df
```

```python
full_pop_df = full_pop_df.rename(columns={'ANO': 'YEAR', 'POPULACAO': 'POPULATION'})
full_pop_df
```

```python
corrected_ufs_df
```

```python
tot_cases = corrected_ufs_df[['ID_MUNICIP']].groupby(['ID_MUNICIP']).size().reset_index(name='TOTAL_CASES').sort_values(by='TOTAL_CASES', ascending=False)
tot_cases
```

```python
# tot_cases = corrected_ufs_df[['ID_MUNICIP']].groupby(['ID_MUNICIP', 'NU_ANO']).size().reset_index(name='TOTAL_CASES').sort_values(by='TOTAL_CASES', ascending=False)

tot_cases_by_year = corrected_ufs_df.groupby(['ID_MUNICIP', 'NU_ANO']).size().reset_index(name='TOTAL_CASES').sort_values(by='TOTAL_CASES', ascending=False)
tot_cases_by_year = tot_cases_by_year.sort_values(by=['NU_ANO', 'ID_MUNICIP'], ascending = True)
tot_cases_by_year = tot_cases_by_year.rename(columns={"NU_ANO": 'YEAR'})
tot_cases_by_year
```

```python
tot_cases_by_year['ID_MUNICIP'] = tot_cases_by_year['ID_MUNICIP'].astype(str)
```

```python
full_pop_df['ID_MUNICIP'] = full_pop_df['ID_MUNICIP'].astype(str)
```

```python
cases_per_year_df = pd.merge(tot_cases_by_year, full_pop_df, on=['ID_MUNICIP', 'YEAR'], how='inner')
cases_per_year_df = cases_per_year_df[['ID_MUNICIP', 'YEAR', 'TOTAL_CASES', 'POPULATION']]
cases_per_year_df
```

```python
annual_incidence_df = cases_per_year_df

annual_incidence_df['CASES_PER_100K'] = (annual_incidence_df['TOTAL_CASES'].astype(int) / annual_incidence_df['POPULATION'].astype(int)) * 100000
annual_incidence_df = annual_incidence_df.sort_values(by='CASES_PER_100K', ascending=False)
annual_incidence_df

### Incidence of cases per municipality per year
```

```python
full_cases_df = pd.merge(tot_cases, full_pop_df, on=['ID_MUNICIP'], how='inner')
full_cases_df
```

```python
mean_population_df = full_cases_df.groupby('ID_MUNICIP').agg({
    'POPULATION': 'mean',
    'TOTAL_CASES': 'first'
}).reset_index()

mean_population_df = mean_population_df.rename(columns={"POPULATION": "MEAN_POPULATION"})
mean_population_df
```

```python
mean_incidence_df = mean_population_df

mean_incidence_df['CASES_PER_100K'] = (mean_incidence_df['TOTAL_CASES'].astype(int) / mean_incidence_df['MEAN_POPULATION'].astype(int)) * 100000
mean_incidence_df = mean_incidence_df.sort_values(by='CASES_PER_100K', ascending=False)
mean_incidence_df

### Mean incidence of cases per municipality across the wole period of study
```
