Getting cases in Manaus by year

```python
tot_cases_manaus_by_year = corrected_ufs_df.groupby(['MUN_RESI', 'YEAR_NOTIF']).size().reset_index(name='TOTAL_CASES').sort_values(by='TOTAL_CASES', ascending=False)
tot_cases_manaus_by_year = tot_cases_manaus_by_year.sort_values(by=['YEAR_NOTIF', 'MUN_RESI'], ascending = True)
tot_cases_manaus_by_year['MUN_RESI'] = tot_cases_manaus_by_year['MUN_RESI'].astype(int)
tot_cases_manaus_by_year['MUN_RESI'] = tot_cases_manaus_by_year['MUN_RESI'].astype(str)
tot_cases_manaus_by_year = tot_cases_manaus_by_year[tot_cases_manaus_by_year.get('MUN_RESI') == '130260']
tot_cases_manaus_by_year
```

Corrects IBGE code to limit to 6 digits

```python
for i in np.arange(0, len(full_pop_df['MUNIC_RES'])):
    if len(full_pop_df['MUNIC_RES'].iloc[i]) > 6:
        full_pop_df['MUNIC_RES'].iloc[i] = full_pop_df['MUNIC_RES'].iloc[i][:-1]
    
full_pop_df
```

```python
full_pop_manaus_df = full_pop_df
full_pop_manaus_df['MUNIC_RES'] = full_pop_manaus_df['MUNIC_RES'].astype(str)
full_pop_manaus_df = full_pop_manaus_df[full_pop_manaus_df.get('MUNIC_RES') == '130260']
full_pop_manaus_df
```

#### Gets number of cases and population per municipality per year

```python
cases_per_year_df = pd.merge(tot_cases_manaus_by_year, full_pop_manaus_df, on=['MUN_RESI', 'YEAR'], how='inner')
cases_per_year_df = cases_per_year_df[['MUN_RESI', 'YEAR', 'TOTAL_CASES', 'POPULATION']]
cases_per_year_df
```

#### Gets annual incidence per municipality per year

```python
annual_incidence_df = cases_per_year_df

annual_incidence_df['CASES_PER_100K'] = (annual_incidence_df['TOTAL_CASES'].astype(int) / annual_incidence_df['POPULATION'].astype(int)) * 100000
annual_incidence_df = annual_incidence_df.sort_values(by='CASES_PER_100K', ascending=False)
annual_incidence_df
```

#### Gets total number of cases in each municipality from 2003 to 2023

```python
tot_cases_manaus = tot_cases_brazil[tot_cases_brazil.get('MUN_RESI') == '130260']
tot_cases_manaus
```

```python
full_cases_df = pd.merge(tot_cases_manaus, full_pop_manaus_df, on=['MUN_RESI'], how='inner')
full_cases_df
```

#### Mean population of Manaus from 2017 to 2021

```python
mean_population_df = full_cases_df.groupby('MUN_RESI').agg({
    'POPULATION': 'mean',
    'TOTAL_CASES': 'first'
}).reset_index()

mean_population_df = mean_population_df.rename(columns={"POPULATION": "MEAN_POPULATION"})
mean_population_df
```

#### Mean incidence of cases in Manaus from 2017 to 2021

```python
mean_incidence_df = mean_population_df

mean_incidence_df['CASES_PER_100K'] = (mean_incidence_df['TOTAL_CASES'].astype(int) / mean_incidence_df['MEAN_POPULATION'].astype(int)) * 100000
mean_incidence_df = mean_incidence_df.sort_values(by='CASES_PER_100K', ascending=False)
mean_incidence_df
```

```python
manaus_cases_per_year_df = cases_per_year_df
manaus_cases_per_year_df
```

```python
manaus_rural_pop_df = manaus_cases_per_year_df.assign(RURAL_POP = manaus_cases_per_year_df.get('POPULATION')*0.45/100)
manaus_rural_pop_df['RURAL_POP'] = manaus_rural_pop_df['RURAL_POP'].round()
manaus_rural_pop_df = manaus_rural_pop_df.assign(RURAL_CASES = manaus_rural_pop_df.get('TOTAL_CASES')*35.75/100)

manaus_rural_pop_df
```

```python
manaus_rural_pop_df['RURAL_INCIDENCE'] = (manaus_rural_pop_df['RURAL_CASES'].astype(int) / manaus_rural_pop_df['RURAL_POP'].astype(int)) * 100000
#manaus_rural_pop_df = manaus_rural_pop_df.sort_values(by='CASES_PER_100K', ascending=False)
manaus_rural_pop_df
```

```python
manaus_pop_difference_df = manaus_rural_pop_df

manaus_pop_difference_df['POPULATION_DIFF'] = manaus_pop_difference_df['POPULATION'] - manaus_pop_difference_df['POPULATION'].shift(1)
manaus_pop_difference_df['POPULATION_DIFF'].iloc[0] = 0

manaus_pop_difference_df['RURAL_POP_DIFF'] = manaus_pop_difference_df['RURAL_POP'] - manaus_pop_difference_df['RURAL_POP'].shift(1)
manaus_pop_difference_df['RURAL_POP_DIFF'].iloc[0] = 0

manaus_pop_difference_df
```

Total increase in population between 2017 and 2021:

```python
manaus_pop_difference_df.get('RURAL_POP').iloc[-1]-manaus_pop_difference_df.get('RURAL_POP').iloc[0]
```

Approximating a birth rate between first and last years:

```python
(manaus_pop_difference_df.get('RURAL_POP').iloc[-1]-manaus_pop_difference_df.get('RURAL_POP').iloc[0])/(len(manaus_pop_difference_df)-1)
```

About 142 births per year from 2017 to 2021

Getting a mean rural population

```python
mean_rural_pop = manaus_pop_difference_df.loc[0:4, 'RURAL_POP'].mean()
mean_rural_pop
```

Around 9840 people living in the rural areas of Manaus

```python
manaus_rural_births_per_day = (manaus_pop_difference_df.get('RURAL_POP').iloc[-1]-manaus_pop_difference_df.get('RURAL_POP').iloc[0])/(len(manaus_pop_difference_df)-1)/365
manaus_rural_births_per_day
```

Around 0.388 births per day

```python
manaus_rural_births_per_person_day = manaus_rural_births_per_day/mean_rural_pop
manaus_rural_births_per_person_day
```
