#### Gets population data for all brazillian municipalities from 2003 to 2023

```python
pop_df = pd.read_csv('pop_data/pop_2003-2023.csv')
pop_df = pop_df.drop(columns=['Unnamed: 0'])
pop_df
```

```python
full_pop_df = pop_df.rename(columns={'ANO': 'YEAR', 'POPULACAO': 'POPULATION'})
full_pop_df
```

#### Total cases by municipality of residence 

```python
tot_cases_brazil = corrected_ufs_df[['MUN_RESI']].groupby(['MUN_RESI']).size().reset_index(name='TOTAL_CASES').sort_values(by='TOTAL_CASES', ascending=False)
tot_cases_brazil['MUN_RESI'] = tot_cases_brazil['MUN_RESI'].astype(int)
tot_cases_brazil['MUN_RESI'] = tot_cases_brazil['MUN_RESI'].astype(str)
tot_cases_brazil
```

Adding an year column for date of notification

```python
corrected_ufs_df['YEAR_NOTIF'] = corrected_ufs_df['DT_NOTIF'].str[-2:].astype(int)
corrected_ufs_df['YEAR_NOTIF'] = corrected_ufs_df['YEAR_NOTIF'].apply(lambda x: 2000 + x if x <= 23 else 1900 + x)
corrected_ufs_df
```

```python
corrected_ufs_df['RES_EXAM'].unique()
```

Getting cases by municipality by year

```python
tot_cases_brazil_by_year = corrected_ufs_df.groupby(['MUN_RESI', 'YEAR_NOTIF']).size().reset_index(name='TOTAL_CASES').sort_values(by='TOTAL_CASES', ascending=False)
tot_cases_brazil_by_year = tot_cases_brazil_by_year.sort_values(by=['YEAR_NOTIF', 'MUN_RESI'], ascending = True)
tot_cases_brazil_by_year
```

```python
tot_cases_brazil_by_year['MUN_RESI'] = tot_cases_brazil_by_year['MUN_RESI'].astype(int)
tot_cases_brazil_by_year['MUN_RESI'] = tot_cases_brazil_by_year['MUN_RESI'].astype(str)
tot_cases_brazil_by_year
```

```python
full_pop_df['MUNIC_RES'] = full_pop_df['MUNIC_RES'].astype(str)
full_pop_df
```

Corrects IBGE code to limit to 6 digits

```python
for i in np.arange(0, len(full_pop_df['MUNIC_RES'])):
    if len(full_pop_df['MUNIC_RES'].iloc[i]) > 6:
        full_pop_df['MUNIC_RES'].iloc[i] = full_pop_df['MUNIC_RES'].iloc[i][:-1]
    
full_pop_df
```

```python
tot_cases_brazil_by_year = tot_cases_brazil_by_year.rename(columns={"YEAR_NOTIF": "YEAR"})
full_pop_df = full_pop_df.rename(columns={"MUNIC_RES": "MUN_RESI"})
```

#### Gets number of cases and population per municipality per year

```python
cases_per_year_df = pd.merge(tot_cases_brazil_by_year, full_pop_df, on=['MUN_RESI', 'YEAR'], how='inner')
cases_per_year_df = cases_per_year_df[['MUN_RESI', 'YEAR', 'TOTAL_CASES', 'POPULATION']]
cases_per_year_df
```

#### Gets annual incidence per municipality per year

```python
annual_incidence_df = cases_per_year_df

annual_incidence_df['CASES_PER_100K'] = (annual_incidence_df['TOTAL_CASES'].astype(int) / annual_incidence_df['POPULATION'].astype(int)) * 100000
annual_incidence_df = annual_incidence_df.sort_values(by='CASES_PER_100K', ascending=False)
annual_incidence_df
```

#### Gets total number of cases in each municipality from 2003 to 2023

```python
full_cases_df = pd.merge(tot_cases_brazil, full_pop_df, on=['MUN_RESI'], how='inner')
full_cases_df
```

#### Mean population of each municipality from 2003 to 2023

```python
mean_population_df = full_cases_df.groupby('MUN_RESI').agg({
    'POPULATION': 'mean',
    'TOTAL_CASES': 'first'
}).reset_index()

mean_population_df = mean_population_df.rename(columns={"POPULATION": "MEAN_POPULATION"})
mean_population_df
```

#### Mean incidence of cases per municipality from 2003 to 2023

```python
mean_incidence_df = mean_population_df

mean_incidence_df['CASES_PER_100K'] = (mean_incidence_df['TOTAL_CASES'].astype(int) / mean_incidence_df['MEAN_POPULATION'].astype(int)) * 100000
mean_incidence_df = mean_incidence_df.sort_values(by='CASES_PER_100K', ascending=False)
mean_incidence_df
```
