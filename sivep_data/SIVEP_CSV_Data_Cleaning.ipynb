Removing unused columns and selecting only cases caused by P. Vivax:

```python
for year in range(3, 24):
    year_str = f"{year:02d}"
    
    malaria_df = pd.read_csv(f'malaria_notification_data_csv/NOTIPO{year_str}.csv')
    
    malaria_df_drop_cols = malaria_df.drop(columns=columns_to_drop, errors='ignore')
    
    malaria_df_drop_cols  = malaria_df_drop_cols[malaria_df_drop_cols.get('RES_EXAM')==4]
    
    malaria_df_drop_cols  = malaria_df_drop_cols[malaria_df_drop_cols.get('PAIS_INF')==1]
    
    malaria_df_drop_cols  = malaria_df_drop_cols[malaria_df_drop_cols.get('PAIS_RES')==1]
    
    malaria_df_drop_cols = malaria_df_drop_cols.drop(columns=['RES_EXAM', 'PAIS_INF', 'PAIS_RES'])
    
    malaria_df_drop_cols.to_csv(f'malaria_notification_data_csv/malaria_vivax_20{year_str}_treated.csv', index=False)
```
